Using Pandas

```python
# check the data
import pandas as pd
df = pd.read_csv('arrhythmia.data', header=None)
```

```python
# FOREWORD:
# Cel: Wybrać z df tylko pierwsze 6 kolumn i nadać im sensowne nazwy zgodne z dokumentacją danych. #
# Wejścia: df (DataFrame z kolumnami indeksowanymi liczbami 0..). #
# Wyjścia: data (DataFrame z 6 kolumnami: age, sex, height, weight, QRS duration, P-R interval). #
# Założenia: df ma co najmniej 6 kolumn; kolumny 0..5 odpowiadają tym cechom wg dokumentacji. #
# Pułapki: jeśli df ma już nazwy kolumn (nie int), df[[0..5]] może nie zadziałać; jeśli kolejność kolumn inna niż w dokumentacji → błędne mapowanie. #

# since the data has many columns, take just the first few and name them (as per the documentation) #
data = df[[0, 1, 2, 3, 4, 5]]  # wybieramy 6 pierwszych cech do dalszej analizy (mniej szumu, łatwiejsza kontrola) #
data.columns = ['age', 'sex', 'height', 'weight', 'QRS duration', 'P-R interval']  # nadajemy nazwy cech, żeby kod był czytelny i spójny z opisem danych #
```
